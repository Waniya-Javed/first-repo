```python
import tkinter as tk
from tkinter import messagebox

accounts = {}

def create_account(name, acc_no ,balance):
    if acc_no in accounts:
        return False
    accounts[acc_no] = {'name' : name , 'balance' : balance}
    return True

def deposit(acc_no,amount):
    if acc_no in accounts:
        accounts[acc_no]['balance'] += amount
        return True
    return False

def withdraw(acc_no,amount):
    if acc_no in accounts and accounts[acc_no]['balance'] >= amount:
        accounts[acc_no]['balance'] -= amount
        return True
    return False 

def get_balance(acc_no):
    return accounts.get(acc_no , {}).get('balance',None)

root = tk.Tk()
root.title("Banking Management System")
root.geometry("500x300")


def open_create_acc():
    window = tk.Toplevel(root)
    window.title("Create Account")

    tk.Label(window, text="Name:").grid(row = 0 ,column = 0)
    tk.Label(window, text="Account No:").grid(row = 1 , column = 0)
    tk.Label(window, text="Initial Balance:").grid(row = 2 , column = 0)

    name_entry = tk.Entry(window)
    acc_entry = tk.Entry(window)
    bal_entry = tk.Entry(window)

    name_entry.grid(row = 0,column = 1)
    acc_entry.grid(row = 1 , column = 1)
    bal_entry.grid(row = 2 , column = 1)

    def submit():
        name = name_entry.get()
        acc = acc_entry.get()
        try:
            balance = float(bal_entry.get())
        except ValueError:
            messagebox.showerror("Error!" ,"Invalid Balance")
            return
        if create_account(name, acc ,balance):
            messagebox.showinfo("Success","Account Created Successfully!")
            window.destroy()
        else:
            messagebox.showinfo("Error!","Account already Exists.")

    tk.Button(window, text="Create", command=submit).grid(row=3, columnspan=2)    

def open_deposit():
    window = tk.Toplevel(root)
    window.title("Deposit")

    tk.Label(window , text = "Account No:").grid(row=0 , column= 0)
    tk.Label(window , text = "Amount:").grid(row = 1 , column = 0)

    acc_entry = tk.Entry(window)
    amt_entry = tk.Entry(window)

    acc_entry.grid(row = 0 , column = 1)
    amt_entry.grid(row = 1 , column = 1)

    def submit():
        acc = acc_entry.get()
        try:
            amount = float(amt_entry.get())
        except ValueError:
            messagebox.showerror("Error!","Invalid Amount.")
            return
        if deposit(acc , amount):
            messagebox.showinfo("Success!", "Amount Deposited.")
            window.destroy()
        else:
            messagebox.showinfo("Error!","Account already Exists.")

    tk.Button(window , text = "Deposit" , command = submit).grid(row = 2 , columnspan = 2)
    
def open_withdraw():
    window = tk.Toplevel(root)
    window.title("Withdraw")

    tk.Label(window , text = "Account No:").grid(row=0 , column= 0)
    tk.Label(window , text = "Amount:").grid(row = 1 , column = 0)

    acc_entry = tk.Entry(window)
    amt_entry = tk.Entry(window)

    acc_entry.grid(row = 0 , column = 1)
    amt_entry.grid(row = 1 , column = 1)
    
    def submit():
        acc = acc_entry.get()
        try:
            amount = float(amt_entry.get())
        except ValueError:
            messagebox.showerror("Error!","Invalid Amount.")
            return
        if withdraw(acc , amount):
            messagebox.showinfo("Success!", "Amount Withdrawn.")
            window.destroy()
        else:
            messagebox.showinfo("Error!","Account already Exists.")

    tk.Button(window , text = "Withdraw" , command = submit).grid(row = 2 , columnspan = 2)

def open_check_balance():
    window = tk.Toplevel(root)
    window.title("Check Balance")

    tk.Label(window , text = "Account No:").grid(row = 0 , column = 0)
    acc_entry = tk.Entry(window)
    acc_entry.grid(row = 0 , column = 1)

    def check():
        acc = acc_entry.get()
        balance = get_balance(acc)
        if balance is not None:
            messagebox.showinfo("Balance" , f"Balance: {balance}")
        else:
            messagebox.showerror("Error!" , "Account not found")
    tk.Button(window , text = "Check Balance" ,command = check).grid(row = 1 , columnspan = 2)


tk.Button(root, text="Create Account", width=20, command=open_create_acc).pack(pady=10)
tk.Button(root, text="Deposit" , width = 20, command=open_deposit).pack(pady=10)
tk.Button(root, text="Withdraw" , width=20 , command=open_withdraw).pack(pady=10)
tk.Button(root, text="Check Balance" , width=20 , command=open_check_balance).pack(pady=10)

root.mainloop()
```
